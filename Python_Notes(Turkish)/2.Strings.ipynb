partition(separator)

String’i, verilen bir ayırıcıya göre üç parçaya böler: tuple(öncesi, ayırıcı, sonrası).


rpartition(separator)

String’i, verilen bir ayırıcıya göre sağdan üç parçaya böler.

```python
s = "hello world"
print(s.partition(" "))  # Çıktı: ('hello', ' ', 'world')
```

```python
s = "hello world"
print(s.rpartition(" "))  # Çıktı: ('hello', ' ', 'world')
```

splitlines(keepends=False)

Satırları bölerek bir liste döner. True alırsa /n de ekler

```python
s = "apple,banana,orange"
s.rsplit(",", 1)
```

```python
s = "apple,banana,orange"
s.rsplit(",")
```

```python
s = "a,b,c"
print(s.split(","))  # Çıktı: ['a', 'b', 'c']
```

```python
s = "hello\nworld"
print(s.splitlines())  # Çıktı: ['hello', 'world']
```

```python
s ="""hello
world"""
print(s.splitlines())  # Çıktı: ['hello', 'world']
```
